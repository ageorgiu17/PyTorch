# Turning Data into Tensors

În lecția precedentă am văzut cum putem să ne creem un set de date utilizând o metodă din Scikit-Learn și cum anume putem vizualiza datele respective. O să refacem din nou acest lucru

````python
# importing the libraries
from sklearn.datasets import make_circles
import seaborn as sns
import pandas as pd
import torch
````

````python
# creating the dataset
# creating the Features and the labels
X, y = make_circles(n_samples=1000, noise=0.03, random_state=42)
````

````python
# creating a DataFrame with the data to be able to visualize better the data
df = pd.DataFrame({"X1": X[:, 0],
                   "X2": X[:, 1],
                   "labels": y})
df.head()
````

````python
# create a scatterplot to visualize the data
sns.scatterplot(data=df, x="X1", y="X2", hue="labels")
````

La un model de Deep Learning din PyTorch trebuie să îi oferim datele de input ca și tensor. O să verificăm acuma ce tip de date au variabilele de X și y cu care o să lucrăm

````python
type(X), type(y)
````

Datele respective sunt prezente momentan în array-uri de Numpy, ceea ce nu putem să folosim ca și date de input pentru un model. Trebuie să transformăm aceste date în tensori, iar în prima secțiune a acestui curs am învățat că putem să facem asta prin utilizarea metode `from_numpy()` prin care se transformă un array din Numpy într-un tensor. În continuare o să facem asta

````python
X = torch.from_numpy(X).type(torch.float32)
y = torch.from_numpy(y).type(torch.float32)
````

Un array din Numpy are ca și tip de date stocat default float64. Modul default prin care lucrează PyTorch este float32, din acest motiv atunci când trebuie să transformăm din Numpy în torch.Tensor trebuie să modificăm și tipul de date, să transformăm din float64 în float32. Acum că am modificat tipul de date, putem să verificăm să vedem dacă totul este corect acuma, dacă totul este cum ne-am așteptat

````python
type(X), type(y)
````

````python
X.dtype, y.dtype
````

Atât tipul variabilelelor cât și datele din cadrul acestora au fost transformate în datele de care avem nevoie pentru a putea lucra mai departe cu acestea

## Recapitulare

În cadrul acestei lecții am recapitulat cum putem să transformăm un tip de date din Numpy array în torch.Tensor
