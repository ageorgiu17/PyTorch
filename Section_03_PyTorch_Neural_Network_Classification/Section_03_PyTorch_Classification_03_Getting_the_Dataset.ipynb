# Creating a Dataset

Primul pas în cadrul unui proiect de Deep Learning este să avem un set de date cu care să lucrăm. O să ne folosim de un set de date din Scikit-Learn, un set de date ce conțin informații despre anumite jucării. Aceste date o să le preluăm din modului `sklearn.datasets`, iar setul de date poartă denumirea de `make_circles`

````python
from sklearn.datasets import make_circles
````

````python
make_circles
````

`make_circles()` este defapt o funcție prin care putem să creem un set de date cu un număr de sample-uri pe care îl alegem atunci când apelăm funcția respectivă. Funcția returnează două seturi de date, unul pentru X și unul pentru y (Features and labels). De asemenea o să creem acest set de date și cu un noise în acest set de date (ca să fie datele mai interesante și mai apropiate de datele din lumea reală, date care mereu o să vină cu un noise în ele) și de asemenea un random seed (**random_state**) ca să putem reproduce acest experiment

````python
X, y = make_circles(n_samples=1000, noise=0.03, random_state=42)
````

````python
len(X), len(y)
````

````python
X[:5], y[:5]
````

Avantajul pe care îl avem când ne creem un astfel de set de date este faptul că acestea vin direct în forma numerică de care avem nevoie (deși poate fi și un dezanvataj deoarece este esențial să șțim cum putem să transformăm datele în valori numerice, în tensori). Datele cu care lucrăm sunt împărțite în Features (X) și labels (y). În funcție de valorile din Features se determină valoarea pentru labels. Acest lucrul am putea să  îl vizualizăm și mai ușor utilizând Pandas

````python
import pandas as pd
````

Acum cu Pandas o să creem un DataFrame unde o să avem Features și labels. În setul curent de date lucrăm cu două features, și anumele cele două valori pentru fiecare element din array

````python
df = pd.DataFrame({'X1': X[:, 0],
                   'X2': X[:, 1],
                   'labels': y})
df
````

În DataFrame-ul de mai sus putem să vizualizăm mai ușor cum anume arată date. Pentru fiecare X1 și X2 din acel DataFrame îi corespunde o valoare din 'labels'. Modelul nostru final asta o să încerce să facă, având datele pentru X1 și X2 trebuie să estimeze cât mai bine valoarea pentru labels (adică 1 sau 0). Putem să vizualizăm și mai bine aceste date utilizând librăria Seaborn print-un scatterplot

````python
import seaborn as sns
````

````python
sns.scatterplot(data=df, x='X1', y='X2', hue='labels')
````

Datele respective sunt așezate sub formă de cerc (de unde și denumirea de **make_circle()**). Din aceste date, modelul pe care o să îl creem trebuie să găsească anumite pattern-uri ca să descopere cele mai bune valori pentru parametrii, astfel încât atunci când îi oferim două valori pentru X1 și X2 modelul să ne spună din care categorie face parte acel element.
