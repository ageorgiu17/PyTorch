# Non-Linear Activation

După cum spunea, pentru ca un model să fie cât mai bun, de cele mai multe ori acesta o să fie construit din mai multe layere combinate (liniare, non-liniare și altele). Am lucrat până în acest moment cel mai mult cu layere liniare, iar acestea reprezintă în mare parte o linie dreaptă. În continuare o să ne creem un tensor care o să fie liniar, iar asupa acestui tensor o să aplicăm unele metode de activare non-liniare pentru a vedea cum anume se modifică acele date.

```python
# importing the libraries
import torch
from torch import nn

import matplotlib.pyplot as plt
```

```python
# creating the linear tensor
X = torch.arange(-10, 11, 1.)
```

```python
plt.plot(X)
```

După cum se poate observa din graficul de mai sus, așa arată un layer liniar, practic este o linie dreaptă fără întrerupere. În continuare o să facem plot la două dintre cele mai utilizate metode de activare pentru un layer non-liniar, și anume:

1. `torch.relu()`

2. `torch.sigmoid()`

```python
plt.plot(torch.relu(X))
```

După se poate observa, prin acestă funcție de activare linia dreaptă care exista înainte acuma s-a modificat, devenind o linie curbă. Această curbă nu este una extrem de smooth, dar prin împărțirea unei linii drepte în mai multe secțiuni se poate crea ușor o linie curbată

```python
plt.plot(torch.sigmoid(X))
```

Cea de a doua funcție de activare pentru un layer non-liniar (`torch.sigmoid()`) returnează o linie curbă mult mai smooth. Acestea două metode sunt cele mai des utilizate în combinație cu layere liniare pentru a se ajunge la rezultatul dorit. Cu combinația aceasta dintre un număr infinit de linii drepte și linii curbe un model de Deep Learning ar putea să învețe să fie aplicat pentru orice problemă. Combinația aceasta de layere liniare și non liniare face ca un model să fie mai potrivit pentru un anumit set de date.
